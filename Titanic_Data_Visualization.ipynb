```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Praktek Python/titanic.csv")
df.head(10)
```

```python
df.info()
```

```python
for col in ['Survived', 'Sex']:
  plt.figure(figsize=(4,4))
  plt.pie(df[col].value_counts(), labels=df[col].unique(), autopct='%1.1f%%', shadow=True, startangle=90)
  plt.axis('equal')
  plt.title(f'percentage by {col}')

  plt.show()
```

Pada pie chart di atas diketahui persentase penumpang yang tidak selamat (0) lebih tinggi dibandingkan selamat (1) dan jenis kelamin berhubungan dengan survived penumpang

```python
sns.countplot(x='Survived', hue="Sex", data=df)
```

Pada bar chart di atas diketahui mayoritas penumpang yang tidak selamat berjeniskelamin Laki-laki sdgkan penumpang selamat didominasi perempuan

```python
sns.countplot(x='Survived', hue="Pclass", data=df)
```

pada bar chart di atas diketahui penumpang yang tidak selamat lebih banyak dari Pclass kategori 3 atau kelas bawah dan paling banyak selamat dari kelas atas

```python
df['Age'].plot.hist()
```

Pada histogram di atas dapat diketahui bahwa rentang umur penumpang kapal titanic mayoritas 20-40 tahun

```python
sns.countplot(x='SibSp', data = df)
```

Pada bar chart di atas penumpang kapal mayoritas (600 orang) tidak memiliki keluarga, 200 orang memiliki 1 keluarga, dan hanya sedikit yang membawa lebih dari 1 keluarga

```python
for col in ['Age','Pclass', 'SibSp', 'Fare']:
    plt.figure(figsize=(12,5))
    sns.distplot(df[col][df.Survived==0])
    sns.distplot(df[col][df.Survived==1])
    plt.legend(['0','1'])
    plt.show()
```

Berdasarkan grafik di atas dapat diketahui puncak penumpang yang tidak selamat berumur 20 tahunan sedangkan yang selamat berumur 30 tahunan, Pclass dan fare mempengaruhi tingkat survived, Pclass dan fare yang rendah menyebabkan peluang tidak selamat yang tinggi, umur dan jumlah anggota keluarga tidak berpengaruh terhadap survived

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Fare', y='Age',hue='Survived',data=df)
#plt.show()
plt.savefig('plot')
```

Fare berhubungan dengan survived penumpang, semakin rendah fare penumpang maka semakin rendah jumlah selamat

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Fare', y='Pclass',hue='Survived',data=df)
#plt.show()
plt.savefig('plot')
```

Pclass dan fare mempengaruhi survived penumpang, semakin tinggi pclass semakin rendah fare dan menyebabkan angka tidak selamat yang tinggi, sebaliknya fare yang tinggi mendapatkan pclass yang rendah dan angka selamatnya lebih tinggi

```python
cor = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```

Dari heatmap diatas bahwa semakin tinggi korelasi antar baris column dengan nilai mendekati 1 menunjukkan korelasi yang positif artinya fare memiliki peranan penting dalam tingkat survived penumpang, Pclass berkorelasi negatif terhadap survived, jadi semakin tinggi pclass survivednya semakin rendah
